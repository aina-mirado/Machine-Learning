#Diet Problem (Linear Programming)


Mathematical model:

min  ∑ x  
contraints:
  A x ≥ b  
   x ≥ 0

where:
- x = dollars spent on each food
- A = nutrients per dollar
- b = minimum nutrient requirements


```python
import pandas as pd
import cvxpy as cp
import numpy as np
```

```python
df = pd.read_csv("/content/sample_data/diet.csv")

foods = df.iloc[:, 0]
A = df.iloc[:, 1:].values.T
```

```python
b = np.array([
    3, 70, 0.8, 12, 5, 1.8, 2.7, 18, 75
])
```

```python
x = cp.Variable(A.shape[1], nonneg=True)

problem = cp.Problem(
    cp.Minimize(cp.sum(x)),
    [A @ x >= b]
)
problem.solve()
```

```python
print(f"Minimum cost: ${problem.value:.4f}\n")

solution = pd.DataFrame({
    "Food": foods,
    "Dollars": x.value
})

print(solution[solution["Dollars"] > 1e-6].to_string(index=False, float_format="%.2f"))
```
