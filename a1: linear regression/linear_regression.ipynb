```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression as SklearnLinearRegression

df = pd.read_csv('data.csv')

features = ['surface_m2', 'bedrooms', 'has_discount']
X = df[features].copy()
y = df['price'].copy()

X = X.fillna(0).astype(int)
X['surface_m2'] = X['surface_m2'].astype(float)
y = y.astype(float)

pd.concat([X, y], axis=1).to_csv('prepared_data.csv', index=False)
print("Data successfully prepared and saved to 'prepared_data.csv'!")
```

```python
class MyLinearRegression:
    def __init__(self):
        self.w = None 

    def train(self, X, y):
        N = X.shape[0]
        ones = np.ones((N, 1))
        X_tilde = np.append(ones, X, axis=1)
        
        self.w = np.linalg.inv(X_tilde.T @ X_tilde) @ X_tilde.T @ y
        print("Manual model trained ")

    def predict(self, X):
        N = X.shape[0]
        X_tilde = np.append(np.ones((N, 1)), X, axis=1)
        return X_tilde @ self.w
```

```python
X_values = X.values
y_values = y.values

my_model = MyLinearRegression()
my_model.train(X_values, y_values)

sk_model = SklearnLinearRegression()
sk_model.fit(X_values, y_values)

print("Both models have been successfully trained.")
```

```python
def test_and_compare(X_input, y_real, model_manual, model_sklearn):
    preds_manual = model_manual.predict(X_input)
    preds_sklearn = model_sklearn.predict(X_input)
    
    comparison_table = pd.DataFrame({
        'Real Price': y_real,
        'MyModel Prediction': preds_manual,
        'Sklearn Prediction': preds_sklearn,
        'Difference (Manual vs Sk)': np.abs(preds_manual - preds_sklearn)
    })
    
    return comparison_table

results = test_and_compare(X_values, y_values, my_model, sk_model)
print(results.head(10))
```
